Реализуйте классификацию эмоциональной окрашенности текстов при помощи объекта pipeline.

1. Найдите тип задач Sentiment Analysis на huggingface.co
2. Найдите модель для русского языка (примеры: rubert-tiny2..., rubert-base...)

В этом коде используется модель ruBert-base от Sberbank Al для анализа настроений на русском языке.

```python
from transformers import pipeline

# Загрузка модели для анализа настроений на русском языке
model_name = "cointegrated/rubert-tiny-sentiment-balanced"
sentiment_analysis = pipeline("sentiment-analysis", model=model_name)

# Пример использования
texts = [
    "Я люблю программирование!",
    "Сегодня ужасная погода.",
    "Мне нравится эта книга.",
    "Это был обычный день."
]

def classify_sentiment(result):
    label = result['label']
    score = result['score']

    # Классификация на основе меток
    if label == 'positive':
        classification = 'Положительный'
    elif label == 'negative':
        classification = 'Отрицательный'
    else:
        classification = 'Нейтральный'

    return classification, score

results = sentiment_analysis(texts)

# Вывод результатов
for text, result in zip(texts, results):
    classification, score = classify_sentiment(result)
    print(f"Текст: {text}\nЭмоциональная окраска: {classification} с вероятностью {score:.2f}\n")
```
